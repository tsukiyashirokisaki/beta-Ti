```python
import torch
import pandas as pd
import numpy as np
import random
import torch.nn as nn
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
import matplotlib.pyplot as plt
from func import loadMS_class,loadTS
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC,SVR
from joblib import dump, load
import os
import xgboost
import pickle

```

```python
def splitTs():    
    df = pd.read_excel("../data/tensile.xlsx")
    y = df["E (GPa)"].values
    col=['Ti', 'Nb', 'Zr', 'Sn', 'Mo', 'Ta']
    x = df[col].values
    del df
    x_tr,x_te,y_tr,y_te= train_test_split(x, y, test_size=0.2,random_state=0)
    mean = np.mean(x_tr, axis = 0) 
    std = np.std(x_tr, axis = 0)
    for i in range(x_tr.shape[0]):
        for j in range(x_tr.shape[1]):
            if not std[j] == 0 :
                x_tr[i][j] = (x_tr[i][j]- mean[j]) / std[j]
    for i in range(x_te.shape[0]):
        for j in range(x_te.shape[1]):
            if not std[j] == 0 :
                x_te[i][j] = (x_te[i][j]- mean[j]) / std[j]
    y_tr = y_tr
    y_te = y_te
    return x_tr,x_te,y_tr,y_te
def splitMs():
    df = pd.read_excel("../data/stability.xlsx")
    col=['Ti', 'Nb', 'Zr', 'Sn', 'Mo', 'Ta']
    x = df[col].values
    y = df["stability"].values
    x_tr,x_te,y_tr,y_te= train_test_split(x, y, test_size=0.2, random_state=0)
    mean = np.mean(x_tr, axis = 0) 
    std = np.std(x_tr, axis = 0)
    for i in range(x_tr.shape[0]):
        for j in range(x_tr.shape[1]):
            if not std[j] == 0 :
                x_tr[i][j] = (x_tr[i][j]- mean[j]) / std[j]
    for i in range(x_te.shape[0]):
        for j in range(x_te.shape[1]):
            if not std[j] == 0 :
                x_te[i][j] = (x_te[i][j]- mean[j]) / std[j]
    y_tr = y_tr
    y_te = y_te
    return x_tr,x_te,y_tr,y_te
def accuracy(pred,label):
    return np.mean(pred==label)
                
```

```python
TsDataset = splitTs()
MsDataset = splitMs()    
```

```python
## cls
x_tr,x_te,y_tr,y_te = MsDataset
for lr in np.arange(0.001,0.01,0.001):
    for md in np.arange(1,11):
        for n in np.arange(1,31):
            clf = xgboost.XGBClassifier(     
                random_state=0,
                learning_rate=lr,
                max_depth=md,
                n_estimators=n,
                objective= 'binary:logistic',
                eval_metric='mlogloss', 
                use_label_encoder=False)
            clf.fit(x_tr,y_tr)
            acc = np.mean(clf.predict(x_te) == y_te)
            if acc > 0.85:
#                 print(acc)
                dump(clf,'xgb/cls/%.2f_%.3f_%d_%d.joblib'%(acc,lr,md,n))
models = os.listdir("xgb/cls/")
data = []
for i in range(len(models)):
    data.append(models[i].replace(".joblib","").split("_"))
df = pd.DataFrame(data,columns=["Acc","lr","max_depth","#_of_estimators"])
df.to_csv("xgb_cls.csv",index=0)

                
```

```python
## reg
x_tr,x_te,y_tr,y_te = TsDataset
for lr in np.arange(0.05,0.1,0.01):
    for md in np.arange(6,20,2):
        for n in np.arange(20,210,10):
            reg = xgboost.XGBRegressor(     
                random_state=0,
                learning_rate=lr,
                max_depth=md,
                n_estimators=n,
                objective= 'reg:squarederror',
                eval_metric='rmse',)
            reg.fit(x_tr,y_tr)
            mse = np.mean((reg.predict(x_te) - y_te)**2)
            print(mse)
            if mse < 100:
                print(mse)
                dump(reg,'xgb/reg/%.2f_%.3f_%d_%d.joblib'%(mse,lr,md,n))
models = os.listdir("xgb/reg/")
data = []
for i in range(len(models)):
    data.append(models[i].replace(".joblib","").split("_"))
df = pd.DataFrame(data,columns=["MSE","lr","max_depth","#_of_estimators"])
df.to_csv("xgb_reg.csv",index=0)
```

```python
x_tr,x_te,y_tr,y_te = TsDataset
root = "xgb/reg/"
models = os.listdir(root)
pred = []
for model in models[:1]:
    reg = load(root+model)
    pred.append(reg.predict(x_te))
np.save("xgb.npy",reg.predict(x_te))
```
