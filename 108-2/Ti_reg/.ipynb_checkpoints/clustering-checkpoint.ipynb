```python

import torch
import torch.nn.functional as F
import pandas as pd
import numpy as np
import random
import torch.nn as nn
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
```

```python
df=pd.read_excel("../Ti_reg/modulus0922.xlsx")
num_colums = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numerical_columns = list(df.select_dtypes(include=num_colums).columns)
df = df[numerical_columns]
df.drop(labels=['Ref.', 'Unnamed: 0'],axis=1, inplace=True)
drop=[]
for i in range(len(df)):
    now=np.array(df.iloc[i][0:6])
    if max(now)!=now[0]:
        drop.append(i)
   
#df=df.drop(drop)
y = df["E (GPa)"].values
```

```python
col=['Ti', 'Nb', 'Zr', 'Sn', 'Mo', 'Ta']
x = df[col].values
```

```python
from sklearn.cluster import SpectralClustering
```

```python
for i in range(10):
    tsne = TSNE(n_components=2, init='random',random_state=i)
    #pca= PCA(n_components=2,whiten=True)
    tsne.fit(x)
    x_transform=tsne.fit_transform(x,y)
    cm = plt.cm.get_cmap('RdYlBu')
    sc=plt.scatter(x_transform[:,0],x_transform[:,1],c=y,vmin=min(y),vmax=max(y),cmap=cm)
    plt.colorbar(sc)
    plt.show()
```
